```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Carga de archivo**

```python
df = pd.read_csv('https://www.datos.gov.co/api/views/ynam-yc42/rows.csv?accessType=DOWNLOAD')
df
```

```python
df.head(2)
```

```python
df1 = df[['ESTU_TIPODOCUMENTO', 'ESTU_GENERO', 'ESTU_DEPTO_RESIDE', 'ESTU_MCPIO_RESIDE', 'PUNT_MATEMATICAS', 'PUNT_C_NATURALES', 'PUNT_INGLES', 'PUNT_LECTURA_CRITICA', 'PUNT_GLOBAL']]
df1
```

# **Histogramas**



```python
text_dict = {'size':16,'color':'blue'};
plt.figure(figsize=(8,6));
plt.hist(df1['PUNT_MATEMATICAS'],color='red');
plt.xlabel('Puntaje Matemáticas',fontdict=text_dict);
```

```python
text_dict = {'size':16,'color':'blue'};
plt.figure(figsize=(8,6));
plt.hist(df1['PUNT_LECTURA_CRITICA']);
plt.xlabel('Lectura Crítica',fontdict=text_dict);
```

```python
sns.set_style('whitegrid');

text_dict = {'size':16,'color':'blue'};
text_legend = {'size':16}
plt.figure(figsize=(8,6));
plt.hist(df1['PUNT_MATEMATICAS'],color='red',alpha=0.4,label='Matemáticas');
plt.hist(df1['PUNT_LECTURA_CRITICA'],color='blue',alpha=0.4,label='Lect Crítica');
#plt.hist(df1['PUNT_INGLES'],color='skyblue',alpha=0.4,label='Inglés');
plt.xlabel('Puntajes',fontdict=text_dict);
plt.legend(loc='best',prop=text_legend);
```

```python
plt.figure(figsize=(8,6));
sns.scatterplot(data=df1,x='PUNT_MATEMATICAS',y='PUNT_INGLES',hue='ESTU_GENERO');
```

```python
plt.figure(figsize=(8,6));
sns.scatterplot(data=df1,x='PUNT_MATEMATICAS',y='PUNT_C_NATURALES',hue='ESTU_GENERO');
```

```python
sns.jointplot(data=df1, x='PUNT_MATEMATICAS',y='PUNT_C_NATURALES',color='g');
```

```python
sns.jointplot(data=df1, x='PUNT_MATEMATICAS',y='PUNT_C_NATURALES',color='g',kind='hex');
plt.colorbar()
```

**SUBCONJUNTOS**



```python
df_baq = df1[df1['ESTU_MCPIO_RESIDE']=='BARRANQUILLA'];
df_baq
```

```python
plt.figure(figsize=(8,6));
sns.scatterplot(data=df_baq,x='PUNT_MATEMATICAS',y='PUNT_C_NATURALES',hue='ESTU_GENERO');
```

```python
sns.jointplot(data=df_baq, x='PUNT_MATEMATICAS',y='PUNT_INGLES',color='g',kind='kde');
```

```python
colu = ['PUNT_MATEMATICAS','PUNT_INGLES','PUNT_C_NATURALES'];
sns.pairplot(data=df_baq, x_vars=colu, y_vars=colu, hue='ESTU_GENERO',height=4);
```
